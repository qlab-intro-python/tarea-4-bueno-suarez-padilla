## **Mercado de autos usados**

Importa el conjunto de datos `neoauto_20240924.csv`
    

```python
import pandas as pd
import numpy as np
```

```python
base_prop = pd.read_csv("neoauto_20240924.csv", delimiter=",")
```

```python
base_prop
```

1. ¿Cuál es el precio promedio de los autos de cada marca (`item_brand`) publicados en cada subcategoría (`item_category_2`)?

```python
# Solución
precio_promedio = base_prop.groupby(["item_brand", "item_category_2"])["item_price"].mean()
print(precio_promedio)
```

2. ¿Cuántos autos tienen más de 100,000 kilómetros (`item_km`) pero un precio (`item_price`) menor al promedio general?

```python
# Solución
precio_promedio_general = base_prop["item_price"].mean()
autos_usados = base_prop[(base_prop["item_km"] > 100000) & (base_prop["item_price"] < precio_promedio_general)]
cantidad_autos = len(autos_usados)
print(cantidad_autos)
```

3. Encuentra el modelo (`item_name`) más caro y el más barato de cada marca (`item_brand`).

```python
# Solución
modelo_caro = base_prop.loc[base_prop.groupby("item_brand")["item_price"].idxmax()]
modelo_barato = base_prop.loc[base_prop.groupby("item_brand")["item_price"].idxmin()]
print("modelo mas caros:")
print(modelo_caro[["item_price", "item_brand", "item_name"]])

print("modelo mas baratos")
print(modelo_barato[["item_price", "item_brand", "item_name"]])

```

4. Encuentra las tres marcas con mayor número de autos financiados por Santander (`item_financed_by`). 

```python
# Solución
autos_financiados = base_prop[base_prop["item_credit"]].groupby(["item_financed_by", "item_brand"]).size()
marcas_financiadas = autos_financiados.sort_values(ascending=False).head(3)
marcas_financiadas
```

5. Agrupa las publicaciones por tipo de transmisión (`item_transmission`) y calcula la desviación estándar del precio (`item_price`) para cada tipo.  

```python
# Solución
desviacion = base_prop.groupby("item_transmission")["item_price"].std()
desviacion
```

6. ¿Cuál es el kilometraje promedio de las publicaciones con el tag "Premium" y de las publicaciones con el tag "Como nuevo"?

```python
# Solución
premiun = base_prop[base_prop["item_tag"] == "Premium"]
premiun_promedio = premiun["item_km"].mean()
como_nuevo = base_prop[base_prop["item_tag"] == "Como nuevo"]
como_nuevo_promedio = como_nuevo["item_km"].mean()
print(premiun_promedio)
print(como_nuevo_promedio)


```

7. ¿Cuáles son las marcas de vehículas para los que todos sus modelos fabricados en los últimos 5 años tienen al menos 5 publicaciones cada una?

```python
current_year = 2024
recent_cars = base_prop[base_prop['item_year'] >= current_year - 5]
recent_model_counts = recent_cars.groupby(['item_brand', 'item_name']).size().reset_index(name='count')
sufficient_publications = recent_model_counts[recent_model_counts['count'] >= 5]

brands_with_all_models = sufficient_publications.groupby('item_brand')['item_name'].nunique()
total_models_by_brand = recent_cars.groupby('item_brand')['item_name'].nunique()

brands_with_all_models_df = pd.DataFrame({
    'brand': brands_with_all_models.index,
    'all_model_count': brands_with_all_models.values
}).set_index('brand')

total_models_by_brand_df = pd.DataFrame({
    'brand': total_models_by_brand.index,
    'total_model_count': total_models_by_brand.values
}).set_index('brand')

comparison_df = pd.merge(brands_with_all_models_df, total_models_by_brand_df, left_index=True, right_index=True)
valid_brands = comparison_df[comparison_df['all_model_count'] == comparison_df['total_model_count']].index.tolist()


if (len(valid_brands) == 0):
    print("No se encontraron marcas cuyos modelos fabricados en los últimos 5 años tengan al menos 5 publicaciones cada uno.")
else:
    print("\nMarcas cuyos modelos fabricados en los últimos 5 años tienen al menos 5 publicaciones cada uno:")
    print(valid_brands)
```

1. Calcule el PBI per cápita real ajustado por poder de paridad de compra (PPP). Use la variable de PBI que se calcula por gasto y la población. ¿Cuáles son los 10 países top?

```python
data['rgdpe_pc_ppp'] = data['rgdpe'] / data['pop']
latest_data = data.dropna(subset=['rgdpe_pc_ppp']).sort_values('year', ascending=False).drop_duplicates('country')
top_10_countries = latest_data.nlargest(10, 'rgdpe_pc_ppp')[['country', 'rgdpe_pc_ppp']]
top_10_countries
```

2. Genera el ránking de los países con el mayor incremento del PBI per cápita real PPP entre los años 1990 y 2019. ¿Dónde se ubica Perú? Para los países que no tengan información para dichos años, puede utilizar datos provenientes de hasta dos años antes o después

```python
data_1990 = data[(data['year'] >= 1988) & (data['year'] <= 1992)].groupby('country')['rgdpe_pc_ppp'].mean().reset_index()
data_2019 = data[(data['year'] >= 2017) & (data['year'] <= 2021)].groupby('country')['rgdpe_pc_ppp'].mean().reset_index()
increment = pd.merge(data_1990, data_2019, on='country', suffixes=('_1990', '_2019'))
increment['growth'] = (increment['rgdpe_pc_ppp_2019']/increment['rgdpe_pc_ppp_1990'])**(1/29)-1
increment = increment.sort_values('growth', ascending=False)

increment['ranking'] = [i for i in range(1, len(increment)+1)]

peru_rank = increment[increment['country'] == 'Peru']



print("\nRanking de crecimiento del PBI per cápita real PPP (1990-2019):")
print(increment[['ranking', 'country', 'growth']])

print("\nRanking de Perú en crecimiento del PBI per cápita real PPP:")
print(peru_rank[['ranking', 'country', 'growth']])
```

3. Para el periodo 2010-2019, calcule el promedio del índice del capital humano. ¿Qué relación existe entre esta variable y el PBI per cápita real PPP calculado previamente? ¿y el número de horas laboradas al año por los trabajadores?

```python
hc_2010_2019 = data[(data['year'] >= 2010) & (data['year'] <= 2019)].groupby('country')['hc'].mean().reset_index()
rgdpe_ppp_avg = data[(data['year'] >= 2010) & (data['year'] <= 2019)].groupby('country')['rgdpe_pc_ppp'].mean().reset_index()
avh_avg = data[(data['year'] >= 2010) & (data['year'] <= 2019)].groupby('country')['avh'].mean().reset_index()
correlation_hc_pbi = hc_2010_2019['hc'].corr(rgdpe_ppp_avg['rgdpe_pc_ppp'])
correlation_hc_avh = hc_2010_2019['hc'].corr(avh_avg['avh'])

print(f"\nCorrelación entre el índice de capital humano y PBI per cápita real PPP (2010-2019): {correlation_hc_pbi:.2f}")
print(f"Correlación entre el índice de capital humano y horas trabajadas anuales (2010-2019): {correlation_hc_avh:.2f}")
```

7.  Genere una variable que ordene a los países según el porcetaje de formación brutal de capital del 2019. Luego, estable diez grupos a manera de deciles. ¿En qué decil se ubica Péru? Vuelva a calcular usando la información de 1990. ¿El Perú se mantiene en el mismo décil o cambia?

```python
data_2019 = data[data['year'] == 2019].dropna(subset=['csh_i'])
data_2019['capital_rank'] = data_2019['csh_i'].rank(ascending=False)
data_2019['decile'] = pd.qcut(data_2019['capital_rank'], 10, labels=False) + 1
peru_decile_2019 = data_2019[data_2019['country'] == 'Peru']['decile'].values[0]

data_1990 = data[data['year'] == 1990].dropna(subset=['csh_i'])
data_1990['capital_rank'] = data_1990['csh_i'].rank(ascending=False)
data_1990['decile'] = pd.qcut(data_1990['capital_rank'], 10, labels=False) + 1
peru_decile_1990 = data_1990[data_1990['country'] == 'Peru']['decile'].values[0]

# Resultado final del cambio de decil
peru_decile_change = peru_decile_2019 != peru_decile_1990

print("\nDecil de Perú en formación bruta de capital en 2019:", peru_decile_2019)
print("Decil de Perú en formación bruta de capital en 1990:", peru_decile_1990)
print("¿Perú cambió de decil?:", peru_decile_change)
```

8. Filtra todas las transacciones de tipo "Debit" en las que el monto sea mayor a 200, y calcula el monto total de esas transacciones. Además, determina cuántos clientes únicos realizaron esas transacciones.

```python
# Solución
montos = Datos[(Datos["Transaction_Type"] == "Debit") & (Datos["Amount"] > 200)]
print(montos)
montos_total = montos["Amount"].sum()
print("la suma de todos los montos es",montos_total)
cliente_unico = montos["Customer_ID"].nunique()
print("la cantidad de clientes que realizaron las operaciones fueron", cliente_unico)
```

9. Identifica las transacciones con valores nulos en la columna (`Amount`) . Rellena esos valores con la mediana de los montos por tipo de transacción (`Transaction_Type`) y luego calcula el monto total de transacciones por región (`region`) .

```python
# Solución
print(Datos[Datos["Amount"].isnull()])
mediana = Datos.groupby("Transaction_Type")["Amount"].median()
print(mediana)
Datos["Amount"] = Datos.apply(lambda row: mediana[row["Transaction_Type"]] if pd.isnull(row["Amount"]) else row["Amount"], axis=1)
print(Datos)
monto_region = Datos.groupby("Region")["Amount"].sum()
print(monto_region)
```

10. Busca transacciones duplicadas considerando las columnas (`Customer_ID`) y (`Date`). Elimina los duplicados manteniendo solo la primera ocurrencia. Luego, agrupa las transacciones restantes por (`Customer_ID`) y calcula el monto promedio y la cantidad de transacciones por cliente.

```python
# Solución
sin_duplicados = Datos.drop_duplicates(subset=["Customer_ID", "Date"])
agrupacion_promedio = sin_duplicados.groupby("Customer_ID")["Amount"].mean()
agrupacion_cantidad = sin_duplicados.groupby("Customer_ID").size()
print("cantidad de veces que cada cliente hace una transaccion:")
print(agrupacion_cantidad)
print("monto promedio de las transacciones:")
print(agrupacion_promedio)

```
